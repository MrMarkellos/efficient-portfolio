# Ethereum

Put your preference ticker (ETH-USD)

```python
ticker = "ETH-USD"
asset = close[ticker].dropna().to_frame().copy()
asset.rename(columns = {"ETH-USD":"Price"}, inplace = True)
asset
```

```python
asset["P_lag1"] = asset.shift(periods = 1)
asset
```

```python
asset ["P_diff"] = asset.Price.sub(asset.P_lag1)
```

```python
asset.Price.div(asset.P_lag1)-1
```

```python
returns = asset["returns"] = asset.Price.pct_change(periods = 1)
returns
```

Calculate expected returns


```python
expected_returns = returns.mean()
expected_returns = expected_returns*365
print("The expected return is {:.2%} per year.".format(expected_returns))
```
