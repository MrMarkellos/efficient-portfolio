# Libraries

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

# Get historical Data using Yahoo Finance

```python
start = "2020-02-03"
end = "2025-02-03"
```

```python
tickers = [ "CW8.MI", "LYTNOW.SW", "ETHE", "SXRC.MU", "IDTP.L","GC=F", "O"  ]
```

```python
df = yf.download(tickers, start ,end)
df
```

```python
df.to_csv("portfolio_data.csv")
```

## Normalizing Financial Time Series to a Base Value (100)

```python
df.Close
close = df.Close.copy()
```

```python
# First Price 
close.iloc[0,0]
```

```python
close.div(close.iloc[0,0])
```

```python
# First Prices all tickers
close.iloc[0]
```

```python
norm = close.div(close.iloc[0]).mul(100)
norm
```

```python
norm.dropna().plot(figsize = (15,8), fontsize = 13, logy = True)
plt.legend(fontsize = 13)
plt.show()
```
